```python
# Reactions
sheet = wb.get_sheet_by_name("Reactions")
count = 0
for i in range(2, sheet.max_row+1):
    old_id = sheet.cell(row = i, column = 1).value
    if not old_id:
        continue
    
    # Add extra underscore
    new_id = re.sub("EX_[a-zA-Z]+_[DL]$", insert_underscore_r, old_id)
    
    # Add "_e" to end
    if new_id[:3] == "EX_":
        if new_id[-2:] != "_e":
            new_id = new_id +"_e"
            
    if new_id[:3] == "DM_":
        r = model.reactions.get_by_id(old_id)
        m_id = [x.id for x in r.metabolites][0]
        new_id = "DM_" + m_id
            
    if new_id != old_id:
        r = model.reactions.get_by_id(old_id)
        r.id = new_id
        sheet.cell(row = i, column = 1).value = new_id
        print(old_id, new_id, [x.id for x in r.metabolites])
        count += 1
print("Changed {0} reaction ids".format(count))
```

```python
def insert_underscore_r(match):
    substring = match.group(0)
    string_list = substring.rsplit("_", 1)
    return string_list[0] + '__' + string_list[1]

def insert_underscore_m(match):
    substring = match.group(0)
    return substring[0] + '_' + substring[1:]
```

```python
# Metabolites

sheet = wb.get_sheet_by_name("Metabolites")
count = 0
for i in range(2, sheet.max_row+1):
    old_id = sheet.cell(row = i, column = 1).value
    if not old_id:
        continue
    # Add extra underscore
    new_id = re.sub("[a-zA-Z]_[DL]_[ce]$", insert_underscore_m, old_id)
    if new_id != old_id:
        model.metabolites.get_by_id(old_id).id = new_id
        sheet.cell(row = i, column = 1).value = new_id
        print(old_id, new_id)
        count += 1
        
    

print("Changed {0} metabolite ids".format(count))
```
