```python
import cobra
import pandas as pd
#from openpyxl import load_workbook, styles, Workbook
#import re
#import escher
#from escher import Builder
```

```python

# Load model
model_path = "P://git/gem_sco/modelS.xml"
# xl_fn = "P://git/gem_sco/modelS.xlsx"
model = cobra.io.read_sbml_model(model_path)
```

```python
kegg_reaction_list = pd.read_csv("P://SCM/kegg_reactions_sco.csv", header = None)
```

```python
kegg_reaction_list = kegg_reaction_list[0].values.tolist()
```

```python
model_kegg_ids = []
for r in model.reactions:
    try:
        kegg_id = r.annotation["kegg.reaction"]
        if len(kegg_id):
            model_kegg_ids.append(kegg_id)
    except:
        pass

```

```python
print(kegg_reaction_list[:10])
model_kegg_ids.sort()
print(model_kegg_ids[:10])
```

```python
missing = []
for r in kegg_reaction_list:
    if not r in model_kegg_ids:
        missing.append(r)
```
