```python
i = 0
mnx_key = "MetaNetX (MNX) Equation" #metanetx
bigg_key = "BiGG"
for reaction in model.reactions:
    try:
        bigg_id = reaction.annotation["bigg.reaction"]
    except KeyError:
        bigg_id = None
    else:
        if bigg_id:
            continue

    try:
        mnx_id = reaction.annotation["metanetx.reaction"]
    except KeyError:
        continue

    idx = reaction_df[reaction_df[mnx_key] == mnx_id].index.tolist()
    if len(idx):
        print(idx, reaction.id, mnx_id)
    if len(idx) > 1:
        print("Skipping ", reaction.id, len(idx))
    else:
        row = reaction_df.loc[idx]
        bigg = row[bigg_key].tolist()
        try:
            bigg = bigg[0]
        except:
            continue
        if not np.isnan(bigg):
            print(reaction.id, mnx_id, bigg)
        
```
